## Coronavirus descriptive analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from glob import glob as globlin ## The 7bb globlin
directories = globlin('./data/COVID-19 Day Level Cases/*')
directories
```

```python
covid_df = pd.read_csv(directories[2])
```

```python
covid_df.columns
```

```python
beijing_df = covid_df[covid_df['Province/State']=='Beijing']
beijing_df.head()
```
